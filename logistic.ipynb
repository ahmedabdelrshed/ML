## Loading the Data

```python
def load_data(data_directory):
            # directories variable contains the directories names which is 
            # in the data_directory path
            directories = [d for d in os.listdir(data_directory) 
                   if os.path.isdir(os.path.join(data_directory, d))]
            # images and labels lists to load the data
            labels = []
            images = []
            
            #the following for loop iterate over the directories which have been stored above
            for i, d in enumerate(directories):
                label_directory = os.path.join(data_directory, d)
                
                # file names store the image files names which stored in each directory
                # Note to change the condition on your case
                file_names = [os.path.join(label_directory, f) 
                              for f in os.listdir(label_directory) 
                              if f.endswith(".ppm")]
                #the follwing code to load each image and preprocess it
                #we extract the label of the image from the directory name
                for f in file_names:
                    print(f)
                    img = skimage.io.imread(f)
                    img_resized=resize(img,(28,28,3))
                    gray_img=rgb2gray(img_resized)
                    images.append(gray_img.flatten())
                    labels.append(int(d))
            return images, labels
        
```


## Logistic Regression With HOG Features

```python
def load_data(data_directory, num_directories):
    directories = [d for d in os.listdir(data_directory) 
                   if os.path.isdir(os.path.join(data_directory, d))]
    labels = []
    images = []
    
    for i, d in enumerate(directories):
        if i >= num_directories:
            break
        
        label_directory = os.path.join(data_directory, d)
        file_names = [os.path.join(label_directory, f) 
                      for f in os.listdir(label_directory) 
                      if f.endswith(".ppm")]
        
        for f in file_names:
            print(f)
            img = skimage.io.imread(f)
            images.append(img)
            labels.append(int(d))
            
    return images, labels
```

```python
# ROOT_PATH = "D:\Image"
# train_data_directory = os.path.join(ROOT_PATH, "Training") 
# test_data_directory = os.path.join(ROOT_PATH, "Testing")  
train_data_directory = 'Training-1'
test_data_directory  = 'Testing-1'
train_images, train_labels = load_data(train_data_directory,6)
test_images, test_labels = load_data(train_data_directory,6)
print(train_labels)
```

```python
# Function to Preprocess Images for HOG
def preprocess_images_for_hog(images, size=(28, 28)):
    processed_images = []
    for img in images:
        # Resize image
        resized_img = cv2.resize(img, size, interpolation=cv2.INTER_AREA)
        
        # Convert to grayscale if it has three channels
        if resized_img.shape[-1] == 3:
            grayscale_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)
        else:
            grayscale_img = resized_img
        
        processed_images.append(grayscale_img)
    return np.array(processed_images)
```

```python
# Preprocess train and test images
train_images_processed = preprocess_images_for_hog(train_images)
test_images_processed = preprocess_images_for_hog(test_images)
```

```python
from sklearn.linear_model import LogisticRegression
from skimage.feature import hog
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Compute HOG features for Training images
train_hog_features = []
for img in train_images_processed:
    hog_img = hog(img, visualize=False)
    train_hog_features.append(hog_img)
```

```python
# Train logistic regression model

logreg = LogisticRegression(max_iter=1000)

logreg.fit(train_hog_features, train_labels)
```

```python
from sklearn.linear_model import LogisticRegression
from skimage.feature import hog
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


# Compute HOG features for test images
test_hog_features = []
for img in test_images_processed:
    hog_img = hog(img, visualize=False)
    test_hog_features.append(hog_img)
```

```python
# Predict on test set
predictions = logreg.predict(test_hog_features)
# Calculate accuracy
accuracy = accuracy_score(test_labels, predictions)
print("\n Accuracy of logistic regression with HOG features: ",accuracy)
print('\n////////////////////////////////////////////////////////////////////////////////////////// \n')
print("Confusion Matrix as Array \n")
conf_matrix = confusion_matrix(test_labels, predictions)
print(conf_matrix)
```
